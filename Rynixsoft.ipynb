```python
# ==============================
# 🌟 Intern Performance Dataset
# Data Cleaning Pipeline
# ==============================

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

# For colored output
from IPython.display import display
import warnings
warnings.filterwarnings('ignore')

print("\033[1;34m=====================================\033[0m")
print("\033[1;32m  INTERN PERFORMANCE DATA ANALYSIS   \033[0m")
print("\033[1;34m=====================================\033[0m")

# --------------------------------
# 📂 Step 1: Load Dataset
# --------------------------------
print("\n📂 Loading Dataset...\n")

df = pd.read_csv("/content/intern_performance_500_dataset.csv")

print("\033[1;36mDataset Preview:\033[0m")
display(df.head())

# --------------------------------
# 📊 Step 2: Dataset Information
# --------------------------------
print("\n📊 Dataset Information\n")
print(df.info())

# --------------------------------
# ❗ Step 3: Check Missing Values
# --------------------------------
print("\n❗ Missing Values in Dataset\n")

missing_values = df.isnull().sum()

missing_table = pd.DataFrame({
    "Column": missing_values.index,
    "Missing Values": missing_values.values
})

display(missing_table.style
        .background_gradient(cmap="Reds")
        .set_properties(**{'text-align': 'center'}))

# --------------------------------
# 🧹 Step 4: Handle Missing Values
# --------------------------------
print("\n🧹 Cleaning Missing Values...")

# Fill numeric values
df.fillna(df.mean(numeric_only=True), inplace=True)

# Fill categorical values
for col in df.select_dtypes(include='object').columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

print("\033[1;32m✔ Missing Values Handled Successfully\033[0m")

# --------------------------------
# 🔄 Step 5: Encode Categorical Data
# --------------------------------
print("\n🔄 Encoding Categorical Columns...")

le = LabelEncoder()

df['department'] = le.fit_transform(df['department'])
df['final_performance'] = le.fit_transform(df['final_performance'])

print("\033[1;32m✔ Encoding Completed\033[0m")

# --------------------------------
# 📈 Step 6: Cleaned Dataset Preview
# --------------------------------
print("\n📈 Cleaned Dataset Preview\n")

display(df.head().style
        .background_gradient(cmap="Blues")
        .set_caption("Cleaned Intern Performance Dataset"))

# --------------------------------
# 💾 Step 7: Save Clean Dataset
# --------------------------------
df.to_csv("cleaned_intern_dataset.csv", index=False)

print("\n\033[1;35m💾 Cleaned dataset saved as 'cleaned_intern_dataset.csv'\033[0m")

print("\n\033[1;32m🎉 DATA CLEANING COMPLETED SUCCESSFULLY!\033[0m")
```
